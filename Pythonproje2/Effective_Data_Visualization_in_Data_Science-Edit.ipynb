## CODING TASK #1. PLOT PIE CHART USING MATPLOTLIB

Belirli bir portföydeki tüm kripto tahsisini içeren bir Pandas Veri Çerçevesi tanımlayın

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

## CODING TASK #2. PLOT SINGLE & MULTIPLE LINE PLOTS USING MATPLOTLIB

- **Use matplotlib on the Pandas DataFrame to plot the data**

KODLAMA GÖREVİ #2. MATPLOTLIB KULLANILARAK TEK VE ÇOK HATLI ARSALAR
- **crypto_daily_prices'i (BTC, ETH ve LTC) okumak için Pandas read_csv'yi kullanın**
- **Verileri çizmek için Pandas DataFrame'de matplotlib'i kullanın**

```python
investments_df = pd.read_csv('crypto_daily_prices.csv')
investments_df
```

```python
investments_df = investments_df.fillna(0)
investments_df
```

```python
investments_df['Date'] = pd.to_datetime(investments_df['Date'])
investments_df.set_index('Date', inplace=True)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(investments_df.index, investments_df['BTC'], label='BTC', linewidth=2, color="#228B22")
plt.plot(investments_df.index, investments_df['ETH'], label='ETH', linewidth=2, color="#40E0D0")
plt.plot(investments_df.index, investments_df['ADA'], label='ADA', linewidth=2, color="#D2691E")

plt.title('BTC, ETH ve ADA Günlük Fiyatlari')
plt.xlabel('Tarih')
plt.ylabel('Fiyat (USD)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.plot(investments_df['Date'], investments_df['BTC'], label='BTC')
plt.xlabel('Tarih')
plt.ylabel('BTC Fiyati')
plt.title('BTC Fiyat Grafigi')
plt.legend()
plt.show()
```

**PRACTICE OPPORTUNITY #2:**
- **Use subplots attribute to plot separate figures for BTC, ETH, and ADA****BTC, ETH ve ADA için ayrı rakamlar çizmek için subplots özelliğini kullanın**

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10,10))

axes[0].plot(investments_df.index, investments_df['BTC'], label='BTC', linewidth=1,color="#8A2BE2")
axes[0].set_title('BTC Daily Prices')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Price (USD)')
axes[0].grid(True)
axes[0].legend()

axes[1].plot(investments_df.index, investments_df['ETH'], label='ETH', linewidth=2,color="#808000")
axes[1].set_title('ETH Daily Prices')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Price (USD)')
axes[1].legend()

axes[2].plot(investments_df.index, investments_df['ADA'], label='ADA', linewidth=2,color="#A52A2A")
axes[2].set_title('ADA Daily Prices')
axes[2].set_xlabel('Date')
axes[2].set_ylabel('Price (USD)')
axes[2].legend()
plt.tight_layout()
plt.show()
```

## CODING TASK #4. PLOT HISTOGRAMS USING MATPLOTLIB
- **A histogram represents data using bars with various heights**
- **Each bar groups numbers into specific ranges**
- **Taller bars show that more data falls within that specific range**
- **Plot the histogram of the returns**
#Getirilerin histogramını çizin

```python
daily_return_df = pd.read_csv('crypto_daily_returns.csv')
daily_return_df
```

```python
daily_return_df = daily_return_df.fillna(0)
daily_return_df
```

```python
mean_df = round(daily_return_df['BTC'].mean(), 2)
std_dt = round(daily_return_df['BTC'].std(), 2) 
num_bins = 40
plt.figure(figsize=(8,6))
plt.hist(mean_df, num_bins, color='skyblue', edgecolor='black')
plt.hist(std_dt, num_bins, color='red', edgecolor='black')

plt.xlabel('Getiriler')
plt.ylabel('Frekans')
plt.title('Getirilerin Dagilimi')
plt.show()
#aslinda tam anlamadim sanirim boyle- bu cripto da isler yolunda gitmiyor demek mi
```

**PRACTICE OPPORTUNITY #4:**
- **Plot the histogram for BTC and ETH returns using 60 bins. Display all histograms on the same figure***60 kutu kullanarak BTC ve ETH getirilerine ilişkin histogramı çizin. Tüm histogramları aynı şekilde görüntüleyin**


```python
btc_returns = daily_return_df['BTC']
eth_returns = daily_return_df['ETH']
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
axes[0].hist(btc_returns, bins=60, color='#FFC0CB', alpha=0.7, label='BTC Returns')
axes[0].set_title('BTC Returns Histogram')
axes[0].set_xlabel('Return')
axes[0].set_ylabel('Frekans')
axes[0].legend()

axes[1].hist(eth_returns, bins=60, color='#40E0D0', alpha=0.7, label='ETH Returns')
axes[1].set_title('ETH Returns Histogram')
axes[1].set_xlabel('Return')
axes[1].set_ylabel('Frekans')
axes[1].legend()
plt.tight_layout()
plt.show()
```
